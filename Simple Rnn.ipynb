```python
import tensorflow as tf # import tensorflow
from tensorflow.keras import layers, models # import layers and models from keras
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=10000) # load imdb dataset
train_data = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=256) # pad sequences
test_data = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=256) # pad sequences
model=models.Sequential() # create a sequential model
model.add(layers.Embedding(10000, 32)) # add embedding layer
model.add(layers.SimpleRNN(32)) # add simple rnn layer
model.add(layers.Dense(1, activation='sigmoid')) # add dense layer
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc']) # compile model
history = model.fit(train_data, y_train, epochs=10, batch_size=128, validation_split=0.2) # fit model
results = model.evaluate(test_data, y_test) # evaluate model
print('Test accuracy:', results[1]) # print test accuracy
eval_results = model.evaluate(test_data, y_test) # evaluate model
print('Evaluation results:', eval_results) # print evaluation results
display('Simple RNN model summary', model.summary()) # display model summary
```

```python
import pandas as pd # import pandas
import matplotlib.pyplot as plt # import matplotlib
import seaborn as sns # import seaborn
sns.set() # set seaborn style
pd.DataFrame(history.history).plot(figsize=(8, 5)) # plot history
plt.grid(True) # plot grid
plt.gca().set_ylim(0, 1) # set y axis limit
plt.show() # show plot
```
