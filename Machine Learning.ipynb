```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
from mpl_toolkits.mplot3d import Axes3D # 3d plot
import seaborn as sns # data visualization
from sklearn.datasets import load_diabetes # diabetes dataset
from sklearn.linear_model import LinearRegression # linear regression model
from sklearn.metrics import mean_squared_error, r2_score # evaluation metrics
from sklearn.model_selection import train_test_split # train-test split
import warnings # ignore warnings
warnings.filterwarnings('ignore')   # ignore warnings
```

```python
diabetes = load_diabetes() # load diabetes dataset
diabetes_data = diabetes.data  # independent variables
diabetes_target = diabetes.target # dependent variable
```

```python
# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(diabetes_data, diabetes_target, test_size=0.2, random_state=0)
```

```python
# create linear regression object
regression = LinearRegression()
```

```python
# train the model using the training sets
regression.fit(X_train, y_train)
```

```python
# make predictions using the testing set
y_pred = regression.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regression.coef_)
```

```python
# The mean squared error
print('Mean squared error:-', mean_squared_error(y_test, y_pred))
```

```python
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```

```python
# plot outputs
fig = plt.figure(figsize=(10, 7)) # figure size
ax = fig.add_subplot(111, projection='3d') # 3d plot
ax.scatter(X_test[:, 0], X_test[:, 1], y_test, c='r', marker='o') # actual
ax.scatter(X_test[:, 0], X_test[:, 1], y_pred, c='b', marker='^')  # predicted
ax.set_xlabel('X0') # x label
ax.set_ylabel('X1') # y label
ax.set_zlabel('Y') # z label
plt.show() # show plot
```

```python
sns.distplot((y_test-y_pred),bins=50, color='salmon')
mean=round(np.mean(y_test-y_pred),2) # mean of residuals
plt.axvline(mean, color='darkgreen', linestyle='--') # vertical line
plt.title('Residuals')
plt.show() # show plot
```
