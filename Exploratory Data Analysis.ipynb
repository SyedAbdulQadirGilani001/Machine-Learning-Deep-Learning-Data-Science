```python
import pandas as pd # Pandas is a Python library used for working with data sets
import seaborn as sns # Seaborn is a Python data visualization library based on matplotlib
import matplotlib.pyplot as plt # Matplotlib is a Python plotting library
from skimpy import skim # Skim is a Python library for data exploration and cleaning
import warnings # Warnings are used to avoid unnecessary warnings
warnings.filterwarnings('ignore') # Ignore warnings
df = sns.load_dataset('diamonds') # Load the diamonds dataset
print(df.head()) # Print the first 5 rows of the dataset
print(df.describe()) # Print the summary statistics of the dataset
print(df.isnull().sum()) # Print the number of missing values in each column
skim(df) # Print the summary statistics of the dataset
```

```python
corr= df.corr(method='spearman') # Calculate the correlation matrix spearman, which is used for ordinal variables
print(corr) # Print the correlation matrix
sns.heatmap(corr, annot=True, cmap='seismic', vmin=-1, vmax=1) # Plot the correlation matrix
plt.title('Correlation Matrix') # Set the title of the plot
plt.show() # Show the plot
```

```python
df = df.drop(['x', 'y', 'z'], axis=1) # Drop the columns x, y, and z
df.head() # Print the first 5 rows of the dataset
```

```python
df.query('carat > 2.5') # Query the dataset for diamonds with carat greater than 2.5
```

```python
df.query('carat > 2.5 & cut == "Ideal"').head() # Query the dataset for diamonds with carat greater than 2.5 and cut equal to Ideal
```

```python
import ydata_profiling as ydp # YData Profiling is a Python library for data exploration and cleaning
ydp.ProfileReport(df[['carat', 'cut', 'color', 'clarity', 'depth',  'price']],title='Diamonds Dataset', explorative=True, minimal=True) # Generate a report of the dataset
```

```python
df.groupby('cut').mean() # Group the dataset by cut and calculate the mean of each column
```
