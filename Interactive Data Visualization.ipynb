```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # data visualization and styling
import warnings # warning filter
warnings.filterwarnings('ignore') # ignore warnings 
df=sns.load_dataset('iris') # load iris dataset
```

```python
plt.barh(df['species'],df['sepal_length'],color='red') # horizontal bar plot
plt.title('Iris Dataset') # title of plot
plt.xlabel('Sepal Length') # x-axis label
plt.ylabel('Species') # y-axis label
plt.show() # show plot
```

```python
import pandas_profiling # pandas profiling for dataset analysis
import dtale # dtale for interactive data exploration
report = pandas_profiling.ProfileReport(df, title='Pandas Profiling Report', explorative=True)
report.to_file("iris.html") # save report as html file
dtale.show(df, ignore_duplicate=True) # show interactive data exploration
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=df,hue='species',fit_reg=False,palette='Set1') # linear regression plot
plt.title('Iris Dataset') # title of plot
plt.xlabel('Sepal Length') # x-axis label
plt.ylabel('Sepal Width') # y-axis label
plt.show() # show plot
```

```python
sns.heatmap(df.corr(),annot=True,cmap='winter',linewidths=0.5,linecolor='black') # heatmap
plt.title('Iris Dataset') # title of plot
plt.show() # show plot
```
